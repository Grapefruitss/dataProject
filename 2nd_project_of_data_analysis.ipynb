```python
import pandas as pd

oecd_population_df = pd.read_csv('./datas/population_of_oecd.csv')
oecd_population_df.head(5)
```

```python
oecd_population_df.info()
```

```python
oecd_doctors_df = pd.read_csv('./datas/02_oecd_physicians.csv')
oecd_doctors_df = oecd_doctors_df[oecd_doctors_df['Variable'] == 'Practising physicians']
oecd_doctors_df = oecd_doctors_df[oecd_doctors_df['Measure'] == 'Number of persons (head counts)']
oecd_doctors_df.head(20)
```

```python
oecd_doctors_df.info()
```

```python
selected_columns = ['Reference area', 'TIME_PERIOD', 'OBS_VALUE']
oecd_population_df = oecd_population_df[selected_columns]
oecd_population_df = oecd_population_df[(oecd_population_df['TIME_PERIOD']>=2011) & (oecd_population_df['TIME_PERIOD']<=2021)]
oecd_population_df
```

```python
groupby_nations_df = oecd_population_df.groupby('Reference area')
nations = oecd_population_df['Reference area']
nations.unique()
```

```python
selected_columns_2 = ['Country', 'Year', 'Value']
oecd_doctors_df = oecd_doctors_df[selected_columns_2]
groupby_country_df = oecd_doctors_df.groupby('Country')
countrys = oecd_doctors_df['Country']
countrys.unique()
```

```python
doctor_ratio_df = pd.merge(oecd_doctors_df, oecd_population_df, left_on=['Year', 'Country'], right_on=['TIME_PERIOD', 'Reference area'])
doctor_ratio_df['Ratio'] = doctor_ratio_df['Value']/doctor_ratio_df['OBS_VALUE']*1000
grouped_df = doctor_ratio_df.groupby('Country')
```

```python
import matplotlib.pyplot as plt

all_data_df = pd.concat([group for key, group in grouped_df])
yearly_mean_ratio = all_data_df.groupby('TIME_PERIOD')['Ratio'].mean()

plt.figure(figsize=(12, 15))
for key, group in grouped_df:
    if 'Korea' in key:
        plt.plot(group['TIME_PERIOD'], group['Ratio'], linewidth=5, color='blue', marker='o', label=key)
    else:        
        plt.plot(group['TIME_PERIOD'], group['Ratio'], marker='o', label=key)
    plt.text(group['TIME_PERIOD'].iloc[0], group['Ratio'].iloc[0], f'{group["Ratio"].iloc[0]:.2f}', va='bottom')
    plt.text(group['TIME_PERIOD'].iloc[-1], group['Ratio'].iloc[-1], f'{group["Ratio"].iloc[-1]:.2f}', va='bottom')

# 전체 평균 선 추가
plt.plot(yearly_mean_ratio.index, yearly_mean_ratio.values, '-', linewidth=5, color='red', marker='o', label='Yearly Mean Ratio')
first_year = yearly_mean_ratio.index[0]
last_year = yearly_mean_ratio.index[-1]
plt.text(first_year, yearly_mean_ratio.iloc[0], f'{yearly_mean_ratio.iloc[0]:.2f}', va='bottom')
plt.text(last_year, yearly_mean_ratio.iloc[-1], f'{yearly_mean_ratio.iloc[-1]:.2f}', va='bottom')

plt.xticks(ticks=[i for i in range(2011, 2022, 2)], labels=[str(i) for i in range(2011, 2022, 2)])

plt.title('Ratio of doctors')
plt.xlabel('Year')
plt.ylabel('Ratio')
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

oecd_population_projection_df = pd.read_csv('./datas/oecd_population_projection_total.csv')
korea_population_projection_df = oecd_population_projection_df[(oecd_population_projection_df['Country']=='Korea') & (oecd_population_projection_df['Age'] != 'Total')]

korea_population_2024_df = korea_population_projection_df[korea_population_projection_df['Time'] == 2024]
korea_population_2035_df = korea_population_projection_df[korea_population_projection_df['Time'] == 2035]
korea_population_2045_df = korea_population_projection_df[korea_population_projection_df['Time'] == 2045]


# 그래프 그리기
fig, ax = plt.subplots()
bar_width = 0.5
age_labels = korea_population_2024_df['Age']
bar1 = ax.barh(age_labels, -korea_population_2024_df['Value'], align='center', height=bar_width, label='2024')
bar2 = ax.barh(age_labels, korea_population_2035_df['Value'], align='center', height=bar_width, label='2035')

# X축, Y축 라벨링과 타이틀 설정
ax.set_xlabel('Population')
ax.set_ylabel('Age')
ax.set_title('Population Projection for Korea by Age')

ax.set_yticks(range(len(age_labels)))
ax.set_yticklabels(age_labels)

# 범례 추가
ax.legend()
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)

plt.show()
```
