```python
selected_columns = ['Reference area', 'TIME_PERIOD', 'OBS_VALUE']
oecd_population_df = oecd_population_df[selected_columns]
oecd_population_df
```

```python
groupby_nations_df = oecd_population_df.groupby('Reference area')
nations = oecd_population_df['Reference area']
nations.unique()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for key, group in groupby_nations_df:
    plt.plot(group['TIME_PERIOD'], group['OBS_VALUE'], marker='o', label=key)

plt.title('인구수 추이')
plt.xlabel('TIME_PERIOD')
plt.ylabel('OBS_VALUE')
plt.legend()
plt.show()
```
