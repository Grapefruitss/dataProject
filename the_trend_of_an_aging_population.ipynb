```python
import pandas as pd
PACH = "./datas"
df = pd.read_csv(PACH + "/oecd_population_projections.csv", encoding = "utf-8-sig", )
```

```python
selected_columns = ['Country','SEX','TIME','Age','Value']
oecd_population_df = df[selected_columns]
```

```python
oecd_population_df = oecd_population_df[oecd_population_df['Country'] != 'OECD - Total']
oecd_population_df = oecd_population_df[oecd_population_df['SEX'] != 'T']
oecd_population_df = oecd_population_df[oecd_population_df['Age'] != '65 and over']
oecd_population_df = oecd_population_df[oecd_population_df['Age'] != 'Share of 65 and over - elderly']
```

```python
Korea = oecd_population_df[oecd_population_df['Country'] == 'Korea']
```

```python
Korea_W = Korea[Korea['SEX'] == 'W']
Korea_M = Korea[Korea['SEX'] == 'M']
Korea_M
```

```python
import matplotlib.pyplot as plt 
import matplotlib

# 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터프레임
df_male = pd.DataFrame(Korea_M)

# 시각화
plt.figure(figsize=(10, 7))
plt.barh(df_male['TIME'], df_male['Value'], color='skyblue')
plt.xlabel('인구')  # x축 라벨 설정
plt.ylabel('연도')  # y축 라벨 설정
plt.title('남성 인구 피라미드')  # 그래프 제목 설정
plt.grid(True)
plt.show()
```
